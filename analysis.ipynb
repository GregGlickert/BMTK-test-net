```python
import matplotlib.pyplot as plt
from bmtool.bmplot.spikes import raster, plot_firing_rate_distribution
from bmtool.analysis.spikes import load_spikes_to_df, get_population_spike_rate,compute_firing_rate_stats

config = '/home/gjgpb9/BMTK-test-net/simulation_config.json'
network_name = 'network'

df = load_spikes_to_df(spike_file='/home/gjgpb9/BMTK-test-net/output/spikes.h5',config=config,network_name=network_name)

raster(df,config=config,network_name=network_name)
plt.show()
```

```python
spike_rate = get_population_spike_rate(df,config=config,network_name=network_name)
spike_rate.sel(type='smoothed').plot(hue='population')# raw or smoothed are the options
plt.title('Population Spike Rate')
plt.show()
```

```python
pop_stats,indivdual_stats = compute_firing_rate_stats(df, groupby=['pop_name'])
plot_firing_rate_distribution(indivdual_stats,groupby='pop_name',plot_type='violin') # type can be box,swarm,violin or a combo 
plt.show()
```
